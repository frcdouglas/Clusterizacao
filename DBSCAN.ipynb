```python
from sklearn.datasets import load_iris
```

```python
from sklearn.cluster import DBSCAN
import pandas
```

```python
df = load_iris(as_frame = True)['data']
df['type'] = load_iris(as_frame = True)['target']

df
```

```python
df.describe()
```

```python
X = load_iris(as_frame =True)['data']
```

```python
dbscan = DBSCAN(eps=1, min_samples = 5)
```

```python
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
df['cluster'] = dbscan.labels_
```

```python
df.groupby(['type','cluster']).count()
```

```python
# devido a forma de funcionamento do dbscan, ele classificou apenas 2 clusters entre os 3 tipos de iris, isso por que os dados das iris 1 e 2 sÃ£o bem proximos.
```

```python
import plotly.express as px
```

```python
fig = px.scatter(df, x='sepal length (cm)', y='sepal width (cm)', color = 'cluster', title = 'dbscan com eps = 1 e minpts = 5')

fig.show()
```

```python
### agora vamos deixar os dados com mesma media e variancia e utilizar o pca para reduzir o numero de variaveis e testar o dbscan novamente
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
Xscaled = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
```

```python
pca_scaled = PCA(n_components = 2)
```

```python
pca_scaled.fit(Xscaled)
```

```python
print(pca_scaled.explained_variance_ratio_)
```

```python
sum(pca_scaled.explained_variance_ratio_)
```

```python
result_pca_scaled = pandas.DataFrame(pca_scaled.transform(Xscaled))
result_pca_scaled.columns = ['pca1', 'pca2']
```

```python
result_pca_scaled.head()
```

```python
dbscan_pca = DBSCAN(eps = 1, min_samples = 5)
dbscan_pca.fit(Xscaled)
```

```python
result_pca_scaled['cluster'] = dbscan_pca.labels_
```

```python
fig = px.scatter(result_pca_scaled, x='pca1', y='pca2', color = 'cluster', title = 'dbscan com eps = 1 e minpts = 5 com PCA')

fig.show()
```

```python
# observando os resultados, vemos que os daods das iris 1 e 2 ficam muito proximas e pelo dbscan trabalhar com proximidade, ele ira ser falho para estes dados
```
